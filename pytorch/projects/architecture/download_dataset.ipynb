# Displaying Images

```python
import torch
import torchvision
import matplotlib.pyplot as plt
```

```python
!pwd
```

```python
fpath = './arcDataset'
image_loader = torchvision.datasets.ImageFolder(fpath)
```

```python
class_names = {idx: cls for cls, idx in image_loader.class_to_idx.items()}
print(f'Total Images Count: {len(image_loader)}\nTotal Classes Count: {len(class_names)}')
```

```python
class_names
```

```python
figure = plt.figure(figsize=(12, 8))
num_of_images = 6
image_count = len(image_loader)

for i in range(1, num_of_images + 1):
    index = int(torch.randint(low=0, high=image_count, size=(1,))[0])
    plt.subplot(2, 3, i)
    plt.axis('off')
    plt.imshow(image_loader[index][0])
    plt.title(f'{class_names[image_loader[index][1]]}')
```
